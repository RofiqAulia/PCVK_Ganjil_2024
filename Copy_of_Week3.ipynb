```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np

print("Mengubah tingkat kecerahan citra")
print("---------------------------------")

try:
    brightness = int(input("Masukkan nilai kecerahan : "))
except ValueError:
    print("Error, not a number")

# Load the image
original = cv.imread("/content/drive/MyDrive/PCVK/Images/girl.jpg")
if original is None:
    print("Error: Gambar tidak ditemukan atau path salah.")
else:
    brightness_images = np.zeros(original.shape, original.dtype)

    # Akses per pixel
    for y in range(original.shape[0]):
        for x in range(original.shape[1]):
            for c in range(original.shape[2]):
                brightness_images[y, x, c] = np.clip(original[y, x, c] + brightness, 0, 255)

    # Display the original and brightness-adjusted images side by side
    final_frame = cv.hconcat([original, brightness_images])
    cv2_imshow(final_frame)
```

2. Implementasikan transformasi contrast pada Google Colaboratory menggunakan
formula yang terdapat pada bagian Ulasan Teori untuk kontras

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np

print("Mengubah kontras dan tingkat kecerahan citra")
print("--------------------------------------------")

try:
    brightness = int(input("Masukkan tingkat kecerahan [-255 - 255] : "))
    contrast = float(input("Masukkan kontras [1.0 - 3.0] \t\t: "))
except ValueError:
    print("Error, not a number")

# Load the image
original = cv.imread("/content/drive/MyDrive/PCVK/Images/girl.jpg")
if original is None:
    print("Error: Gambar tidak ditemukan atau path salah.")
else:
    # Adjust contrast and brightness
    adjusted_image = cv.convertScaleAbs(original, alpha=contrast, beta=brightness)

    # Display the original and adjusted images side by side
    final_frame = cv.hconcat([original, adjusted_image])
    cv2_imshow(final_frame)
```

3. Implementasikan transformasi logarithmic brightness pada Google Colaboratory
menggunakan formula yang terdapat pada bagian Ulasan Teori untuk transformasi
log

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np

print("Mengubah tingkat kecerahan citra dengan transformasi log")
print("--------------------------------------------")

try:
    brightness = int(input("Masukkan nilai kecerahan : "))
except ValueError:
    print("Error, not a number")

# Load the image
original = cv.imread("/content/drive/MyDrive/PCVK/Images/girl.jpg")
if original is None:
    print("Error: Gambar tidak ditemukan atau path salah.")
else:
    # Adjust the brightness by adding the user input value
    x = original + brightness

    # Apply logarithmic transformation
    c = 255 / np.log(1 + np.max(x))  # Scaling constant
    log_transformed = c * np.log(1 + x)

    # Normalize the output to be within the 0-255 range and convert to uint8
    log_transformed = np.array(log_transformed, dtype=np.uint8)

    # Display the original and adjusted images side by side
    final_frame = cv.hconcat([original, log_transformed])
    cv2_imshow(final_frame)
```

4. Implementasikan transformasi grayscale menggunakan metode averaging, lightness,
dan luminance pada Google Colaboratory menggunakan formula yang terdapat pada
bagian Ulasan Teori

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

print("Menggunakan Averaging")
print("-----------------------------------------")

# Load the image
original = cv.imread("/content/drive/MyDrive/PCVK/Images/ktp-kebumen.jpeg")
if original is None:
    print("Error: Gambar tidak ditemukan atau path salah.")
else:
    # Convert image to float32 for precision in calculations
    original_float = original.astype(np.float32)

    # Apply the averaging formula (R + G + B) / 3
    averaging = np.mean(original_float, axis=2)

    # Convert back to uint8 (8-bit) format
    averaging = averaging.astype(np.uint8)

    # Display each image separately using matplotlib
    plt.figure(figsize=(10, 5))

    # Display original image
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(original, cv.COLOR_BGR2RGB))
    plt.title('Original Image')

    # Display averaged image (grayscale)
    plt.subplot(1, 2, 2)
    plt.imshow(averaging, cmap='gray')
    plt.title('Averaged Image')

    # Show the plot
    plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

print("Menggunakan Lightness")
print("-----------------------------------------")

# Load the image
original = cv.imread("/content/drive/MyDrive/PCVK/Images/ktp-kebumen.jpeg")
if original is None:
    print("Error: Gambar tidak ditemukan atau path salah.")
else:
    # Convert image to float32 for precision in calculations
    original_float = original.astype(np.float32)

    # Calculate the lightness (L) using the formula: (max(R,G,B) + min(R,G,B)) / 2
    lightness = (np.max(original_float, axis=2) + np.min(original_float, axis=2)) / 2

    # Convert back to uint8 (8-bit) format
    lightness = lightness.astype(np.uint8)

    # Display each image separately using matplotlib
    plt.figure(figsize=(10, 5))

    # Display original image
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(original, cv.COLOR_BGR2RGB))
    plt.title('Original Image')

    # Display lightness image (grayscale)
    plt.subplot(1, 2, 2)
    plt.imshow(lightness, cmap='gray')
    plt.title('Lightness Image')

    # Show the plot
    plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

print("Menggunakan Luminance")
print("-----------------------------------------")

# Load the image
original = cv.imread("/content/drive/MyDrive/PCVK/Images/ktp-kebumen.jpeg")
if original is None:
    print("Error: Gambar tidak ditemukan atau path salah.")
else:
    # Convert image to float32 for precision in calculations
    original_float = original.astype(np.float32)

    # Extract R, G, B channels
    B = original_float[:, :, 0]
    G = original_float[:, :, 1]
    R = original_float[:, :, 2]

    # Apply the luminance formula Y' = 0.299 * R + 0.587 * G + 0.114 * B
    luminance = 0.299 * R + 0.587 * G + 0.114 * B

    # Convert back to uint8 (8-bit) format
    luminance = luminance.astype(np.uint8)

    # Display each image separately using matplotlib
    plt.figure(figsize=(10, 5))

    # Display original image
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(original, cv.COLOR_BGR2RGB))
    plt.title('Original Image')

    # Display luminance image (grayscale)
    plt.subplot(1, 2, 2)
    plt.imshow(luminance, cmap='gray')
    plt.title('Luminance Image')

    # Show the plot
    plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

print("Menampilkan Warna Biru dan Mengubah Bagian Lain Menjadi Grayscale")
print("----------------------------------------------------------------")

# Load the image
original = cv.imread("/content/drive/MyDrive/PCVK/Images/ktp-kebumen.jpeg")
if original is None:
    print("Error: Gambar tidak ditemukan atau path salah.")
else:
    # Convert the image to the HSV color space
    hsv_image = cv.cvtColor(original, cv.COLOR_BGR2HSV)

    # Define the lower and upper range for the color blue in HSV
    lower_blue = np.array([100, 150, 0])  # Lower bound for blue hue
    upper_blue = np.array([140, 255, 255])  # Upper bound for blue hue

    # Create a mask to detect blue regions
    mask = cv.inRange(hsv_image, lower_blue, upper_blue)

    # Convert the original image to grayscale
    grayscale = cv.cvtColor(original, cv.COLOR_BGR2GRAY)

    # Convert grayscale back to BGR format
    grayscale_bgr = cv.cvtColor(grayscale, cv.COLOR_GRAY2BGR)

    # Combine the original image and grayscale image using the mask
    result = np.where(mask[:, :, np.newaxis] == 255, original, grayscale_bgr)

    # Display the original image and the result
    plt.figure(figsize=(10, 5))

    # Display original image
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(original, cv.COLOR_BGR2RGB))
    plt.title('Original Image')

    # Display result image
    plt.subplot(1, 2, 2)
    plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))
    plt.title('Blue Highlighted, Rest Grayscale')

    # Show the plot
    plt.show()
```
